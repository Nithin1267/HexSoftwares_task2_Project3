```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------------
# Load Dataset
# -----------------------------------
df = pd.read_csv("C:\\WEB dev\\OneDrive\\Desktop\\ML PROJECT 3\\movies.csv")

# -----------------------------------
# Display Dataset
# -----------------------------------
print("===== FIRST 5 ROWS =====")
print(df.head())

print("\n===== DATASET INFO =====")
df.info()

print("\n===== SHAPE =====")
print(df.shape)

print("\n===== COLUMNS =====")
print(df.columns)

# -----------------------------------
# Data Cleaning
# -----------------------------------
df = df.dropna()
df = df.drop_duplicates()

# Convert rating to numeric (safe)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df.dropna(subset=['rating'])

print("\n===== AFTER CLEANING =====")
print(df.isnull().sum())

# -----------------------------------
# Summary Statistics
# -----------------------------------
print("\n===== SUMMARY STATISTICS =====")

mean_rating = df['rating'].mean()
median_rating = df['rating'].median()
mode_rating = df['rating'].mode()[0]

print("Mean:", mean_rating)
print("Median:", median_rating)
print("Mode:", mode_rating)

# -----------------------------------
# Histogram (Ratings Distribution)
# -----------------------------------
plt.hist(df['rating'], bins=10)
plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()

# -----------------------------------
# Box Plot
# -----------------------------------
plt.boxplot(df['rating'])
plt.title("Box Plot of Ratings")
plt.ylabel("Rating")
plt.show()

# -----------------------------------
# Top Rated Movies
# -----------------------------------
top_movies = df.sort_values(by='rating', ascending=False).head(10)

print("\n===== TOP 10 MOVIES =====")
print(top_movies[['title', 'rating']])

# -----------------------------------
# Top Genres
# -----------------------------------
top_genres = df['genre'].value_counts()

print("\n===== TOP GENRES =====")
print(top_genres)

# -----------------------------------
# Bar Chart (Genres)
# -----------------------------------
top_genres.plot(kind='bar')
plt.title("Top Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()

# -----------------------------------
# Additional Visualization (Optional)
# -----------------------------------
plt.figure(figsize=(6,4))
sns.countplot(x='genre', data=df)
plt.title("Genre Count (Seaborn)")
plt.xticks(rotation=45)
plt.show()

# -----------------------------------
# Conclusion
# -----------------------------------
print("\n===== CONCLUSION =====")
print("""
The movie dataset was cleaned and analyzed.
Mean, median, and mode of ratings were calculated.
Histogram and box plot show rating distribution.
Top movies and genres were identified.
Visualization helped understand trends in movie ratings.
""")
```
